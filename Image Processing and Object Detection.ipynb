```python
def convertG(img):
  rows=img.shape[0]
  cols=img.shape[1]
  abc= np.zeros((rows,cols),dtype=np.uint8)

  for i in range(rows):
    for j in range(cols):
      B=img[i,j,0]//3
      G=img[i,j,1]//3
      R=img[i,j,2]//3

      abc[i,j]= B+G+R
  return abc
```

```python
def binary(image):
  Bimg = np.zeros_like(image)
  for i in range(Bimg.shape[0]):
    for j in range(Bimg.shape[1]):

      if image[i,j]>=100:
        Bimg[i,j]=1
      else:
        Bimg[i,j]=0
  return Bimg
Bimg=binary(Gimg)
```

```python
def find_connected_components(image):
    def dfs(pixel, current_label):
      stack = [pixel]
      while stack:
        y, x = stack.pop()
        if image[y, x] == 1 and labels[y, x] == 0 :
            labels[y, x] = current_label
            for ny, nx in get_adjacent_pixels((y, x)):
              stack.append((ny, nx))
    def is_m_adjacent(p, q):
      V={1}
      # Check if q is in N4(p)
      if abs(p[0] - q[0]) + abs(p[1] - q[1]) == 1:
        return True
      # Check if q is in ND(p) and N4(p)N4(q) has no pixels whose values are from V
      elif abs(p[0] - q[0]) == 1 and abs(p[1] - q[1]) == 1:
        n4_p = [(p[0]+dx, p[1]+dy) for dx, dy in [(0,1), (0,-1), (1,0), (-1,0)]]
        n4_q = [(q[0]+dx, q[1]+dy) for dx, dy in [(0,1), (0,-1), (1,0), (-1,0)]]
        common_neighbors = set(n4_p) & set(n4_q)
        for neighbor in common_neighbors:
          if image[neighbor[0]][neighbor[1]] in V:
            return False
        return True
      else:
          return False


    def get_adjacent_pixels(pixel):
      y, x = pixel
      adjacent_pixels = []
      for j in range(-1, 2):
        for i in range(-1, 2):
          if (i != 0 or j != 0) and 0 <= y+j < height and 0 <= x+i < width:
            if is_m_adjacent((y,x), (y+j,x+i)):
              adjacent_pixels.append((y+j, x+i))
      return adjacent_pixels

    height, width = image.shape
    labels = np.zeros_like(image)
    current_label = 1

    for y in range(height):
        for x in range(width):
             if image[y, x] == 1 and labels[y, x] == 0:
                dfs((y, x), current_label)
                current_label += 1

    num_labels = current_label - 1
    return num_labels, labels


```

```python
num_labels, labels = find_connected_components(Bimg)

print("Number of labels:", num_labels)
print("Labels of connected components:")
print(labels)
```

```python
normalized_labels = labels / np.max(labels)
normalized_labels = normalized_labels.astype(np.float32)
label_hue = np.uint8(179 * normalized_labels)
blank_ch = 255 * np.ones_like(label_hue)
labeled_img = cv2.merge([label_hue, blank_ch, blank_ch])
```

```python
labeled_img = cv2.cvtColor(labeled_img, cv2.COLOR_HSV2BGR)
```
